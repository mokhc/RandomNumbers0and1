```python
factor1 = 200
countlist1 = [(0.5)*factor1, (0.5)*factor1]
```

```python
countlist1 = [int(round(countlist1[0])), int(round(countlist1[1]))]
print(countlist1)
```

```python
# set the mean and standard deviation of the targeted numbers
# in this case, the numbers are 0 and 1
# since the interval for 0 to 1 can be 0.5, we will set the mean based on 0 to 0.5 and 0.5 to 1
# the mean for 0 to 0.5 is 0.25 and the mean for 0.5 to 1 is 0.75
# the standard deviation is used calculated using 128 X values starting from 0 to 128 with intervals of 0.003906
```

```python
meanlist1 = [0.25, 0.75]
stdlist1 = [0.145461, 0.145461]
```

```python
# check the length
```

```python
catlist1 = len(countlist1)
print(catlist1)
```

```python
# function to get normal distribution values based on the meanlist1 and stdlist1
```

```python
def normdist(list1,listindex):
    medianlist1 = []
    alllist1 = []
    num1 = countlist1[listindex]
    index = 0
    mean1 = meanlist1[listindex]
    std1 = stdlist1[listindex]
    for ind1 in range(0, num1):
        value = -1
        medianlist1 = []
        while (value < 0):
            value = st.norm.rvs(loc=mean1, scale=std1, size=1)
            if (value > 0):
                index+=1
                break
        list1.append(value)
        if (index == num1):
            index = 0
            listindex+=1
            medianlist1 = medianlist1 + list1
            alllist1 = alllist1 + list1
    return medianlist1, alllist1
```

```python
# setup variables and convert 500 sets of X values to the median values
```

```python
temp = []
temp2 = []
tempmedian1 = []
tempnumber = []
medianlist1 = []
median200 = []

tempnum = 0
tempall = []
```

```python
listcount=0
# loop 500 times to get 500 sets of values
while (listcount != 500):
    for ind1 in range(catlist1):
        list1 = []
        temp, temp2 = normdist(list1, ind1)
        tempmedian1 = tempmedian1 + temp
        tempall = tempall + temp2
    tempnumber = num.median(tempmedian1)
    if (tempnumber <= 0.5):
        median200.append(0)
    else:
        median200.append(1)
    tempmedian1 = []
    listcount+=1
```
