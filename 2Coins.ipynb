```python
import numpy as num
from scipy import stats as st
import matplotlib.pyplot as mat
import seaborn as sb
import pandas as pan
```

```python
### This piece of code is used generate the scenarios of flipping two coins simultaneously
```

```python
# Determine the sampling number
```

```python
factor1 = 2
countlist1 = [(0.5)*factor1, (0.5)*factor1]
```

```python
countnum = int(countlist1[0] + countlist1[1])
print(countnum)
```

```python
countlist1 = [int(round(countlist1[0])), int(round(countlist1[1]))]
print(countlist1)
```

```python
# set the mean and standard deviation of the targeted numbers
# the result of a coin flip can be thought of as either 0(T) or 1(H)
# the mean would be 0.5
# the standard deviation is used calculated using 256 X values starting from 0 to 256 with intervals of 0.003906
```

```python
meanlist1 = [0.5, 0.5]
stdlist1 = [0.289801, 0.289801]
```

```python
# check the length
```

```python
catlist1 = len(countlist1)
print(catlist1)
```

```python
# function to get normal distribution values based on the meanlist1 and stdlist1
# the values are converted to either 0 or 1
```

```python
def normdist(list1,listindex):
    medianlist1 = []
    alllist1 = []
    value2 = []
    num1 = countlist1[listindex]
    index = 0
    mean1 = meanlist1[listindex]
    std1 = stdlist1[listindex]
    for ind1 in range(0, num1):
        medianlist1 = []
        value = st.norm.rvs(loc=mean1, scale=std1, size=1)
        list1.append(value)
        print("list1 is :",list1)
        if (value <= 0.5):
            value2.append(0)
            index+=1
        else:
            value2.append(1)
            index+=1
        medianlist1 = medianlist1 + value2
        alllist1 = alllist1 + value2
        if (index > num1):
            index = 0
            listindex+=1
    return medianlist1, alllist1
```

```python
# setup variables and collect 1000 sets of X values
```

```python
temp = []
temp2 = []
tempmedian1 = []
tempnumber = []
medianlist1 = []
median1000 = []

tempnum = 0
tempall = []
```

```python
listcount=0
# loop 1000 times to get 1000 sets of values
while (listcount != 1000):
    for ind1 in range(catlist1):
        list1 = []
        temp, temp2 = normdist(list1, ind1)
        tempmedian1 = tempmedian1 + temp
        tempall = tempall + temp2
        print("tempmedian1 is ",tempmedian1)
    tempnumber = tempmedian1
    print("tempnumber is :",tempnumber)
    median1000.append(tempnumber)
    tempmedian1 = []
    listcount+=1
```

```python
print(median1000)
```

```python
# setup a dataframe for the values
```

```python
df = pan.DataFrame(median1000,columns=["C1","C2"])
print(df)
```

```python
# create a new column for categorizing the result of the values
# the values are separated to four different categories depicting the four scenarios: TT, TH, HT, HH. The four scenarios are labeled as 1, 2, 3 and 4.
```

```python
df["Result"] = 0
for ind1 in range(len(df)):
    if (df["C1"][ind1] == 0 and df["C2"][ind1] == 0):
        df["Result"][ind1] = 1
    elif (df["C1"][ind1] == 0 and df["C2"][ind1] == 1):
        df["Result"][ind1] = 2
    elif (df["C1"][ind1] == 1 and df["C2"][ind1] == 0):
        df["Result"][ind1] = 3
    else:
        df["Result"][ind1] = 4
print(df)
```
